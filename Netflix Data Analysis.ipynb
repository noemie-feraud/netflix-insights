```python
import pandas as pd

# Chargement du dataset (remplace 'netflix_titles.csv' par le nom exact de ton fichier)
df = pd.read_csv('netflix_titles.csv')
```

```python
# Question 9 : Calcul du pourcentage de données manquantes
missing_percentage = (df.isnull().sum() / len(df)) * 100
print(missing_percentage)
```

```python
# Question 12 : Toutes les informations de l'oeuvre "Catch me if you can"
df[df['title'] == 'Catch Me If You Can']
```

```python
# On filtre par type 'Movie' puis on trie par année de sortie
df[df['type'] == 'Movie'].sort_values('release_year', ascending=False).head(1)
```

```python
# on filtre par type 'Tv show' puis on trie par année de sortie
df[df['type'] == 'Tv Show'].sort_values('release_year', ascending=False).head(1)
```

```python
# Question 15 : Conversion en type DateTime
df['date_added'] = pd.to_datetime(df['date_added'].str.strip())
```

```python
# étape 1 : on crée la nouvelle colonne en nettoyant le texte de 'duration'
# On retire "min", "Season" et le "s" final de "Seasons"
df['duration_num'] = df['duration'].str.replace(' min', '').str.replace(' Season', '').str.replace('s', '')

# étape 2 : on transforme cette nouvelle colonne en vrais nombres
df['duration_num'] = pd.to_numeric(df['duration_num'], errors='coerce')

# étape 3 : on vérifie que la nouvelle colonne existe et est bien remplie
df[['duration', 'duration_num']].head()
```

```python
# question 18 : transformer la chaîne 'listed_in' en une liste de chaînes
df['listed_in'] = df['listed_in'].str.split(', ')

# vérification du résultat 
df['listed_in'].head()
```

```python
# question 19 : Affichage des valeurs uniques
print("Types d'oeuvres :", df['type'].unique())
print("\nQuelques pays présents :", df['country'].unique()[:10]) # on en affiche 10 pour ne pas saturer l'écran
print("\nAnnées de sortie :", sorted(df['release_year'].unique()))
```

```python
# question 20 : Réalisateurs ayant produit plus d'une oeuvre
directors_counts = df['director'].value_counts()
print(directors_counts[directors_counts > 1])
```

```python
# 1. On s'assure que la colonne date est bien au format DateTime

# 2. Filtrage
movies = df[df['type'] == 'Movie']
shows = df[df['type'] == 'TV Show']

# 3. Extraction de l'année et calcul du record
year_movies = movies['date_added'].dt.year.value_counts()
year_shows = shows['date_added'].dt.year.value_counts()

# 4. Affichage des résultats pour les Questions 21 et 22
print(f"Question 21 - Année record FILMS : {int(year_movies.idxmax())} ({int(year_movies.max())} ajouts)")
print(f"Question 22 - Année record SÉRIES : {int(year_shows.idxmax())} ({int(year_shows.max())} ajouts)")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# On définit le style visuel
sns.set_theme(style="whitegrid")

# Création du graphique (Camembert / Pie Chart)
plt.figure(figsize=(8, 8))
df['type'].value_counts().plot.pie(
    autopct='%1.1f%%', 
    colors=['#E50914', '#221f1f'], # Rouge Netflix et Noir/Gris
    startangle=90,
    explode=(0.05, 0) # On détache un peu la part des Films
)

plt.title('Répartition du catalogue Netflix : Films vs Séries', fontsize=15, fontweight='bold')
plt.ylabel('') # On cache le texte inutile sur le côté
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# On prend les 10 premiers pays 
top_countries = df['country'].value_counts().head(10)

sns.barplot(x=top_countries.values, y=top_countries.index, palette='Reds_r')

plt.title('Top 10 des pays producteurs de contenu', fontsize=15, fontweight='bold')
plt.xlabel('Nombre total d\'oeuvres')
plt.ylabel('Pays')
plt.show()
```

```python
# Question 24 : Comptage des genres (on explose la liste)
genres_exploded = df['listed_in'].explode()

# On compte les occurences de chaque genre unique
top_genres = genres_exploded.value_counts().head(10)

# Affichage pour vérification
print("Top 10 des genres les plus fréquents :")
print(top_genres)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# On définit une taille de figure adaptée
plt.figure(figsize=(12, 8))

# Création du graphique à barres horizontales 
sns.barplot(
    x=top_genres.values,
    y=top_genres.index,
    palette='viridis' # Un dégradé de couleurs professionnel

)

# Personnalisation des labels et du titre
plt.title('Top 10 genres les plus populaires sur Netflix', fontsize=16, fontweight='bold')
plt.xlabel('Nombre d\'oeuvre', fontsize=12)
plt.ylabel('Genre', fontsize=12)

# Affichage du graphique
plt.show()
```
